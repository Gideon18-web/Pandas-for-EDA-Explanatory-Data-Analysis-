## Knowing your data

```python
# Imports
import pandas as pd
```

```python
df = pd.read_csv('health.csv')
df.head()
```

```python
#1st row(index 0) - delete it
df.drop(0,inplace=True)
#Reset index
df = df.reset_index(drop = True)
df.head()
```

## Data Cleaning

```python
#Drop empty columns and rows
df = df.dropna(axis=1).dropna()
df.info()
```

```python
df['CharacteristicLabel'].unique()
```

```python
df['Location'].unique()
```

```python
df=df.drop('CharacteristicLabel',axis=1)
df.head()
```

```python
#Check and drop Columns with 1 value
non_cols = df.columns[df.nunique()==1]
df = df.drop(non_cols,axis=1)
df.head()
```

```python
#Drop columns with 'Id'
id_cols = [x for x in df.columns if 'id' in x.lower()[-2:]]
df.drop(columns=id_cols,inplace=True)
df.head()
```

```python
#Drop columns with Order
order_cols = [x for x in df.columns if 'Order' in x]
df.drop(columns=order_cols,inplace=True)
df.head()
```

```python
#Drop SurveyYearLabel
df.drop(columns='SurveyYearLabel',inplace=True)
df.head()
```

```python
#Change Value to Float | Change Precicion to int
df['Value'] = df['Value'].astype(float)
df['Precision'] = df['Precision'].astype(int)
df.info()
```

##Region by Insurance

```python
#Highest And Lowest Health Insurance Coverages
region_avg = df.groupby(['Location'])['Value'].mean().sort_values(ascending=False)
lowest_region_avg = region_avg.tail(1)
highest_region_avg = region_avg.head(1)

#Find the difference
highest_value = highest_region_avg.values[0]
lowest_value = lowest_region_avg.values[0]
diff = highest_value - lowest_value
perc_diff = (diff/highest_value) * 100

print("Highest Region Coverage:", highest_region_avg.index[0],"with a coverage of:",highest_value)
print("Lowest Region Coverage:", lowest_region_avg.index[0],"with a coverage of:",lowest_value)
print("Percentage Difference Between Highest and Lowest Region:", round(perc_diff,2),"%")
```

```python
#Highest And Lowest Health Insurance Coverages per year
region_2008_avg = df[df['SurveyYear'] == "2008"].groupby(['Location'])['Value'].mean().sort_values(ascending=False)
region_2014_avg = df[df['SurveyYear'] == "2014"].groupby(['Location'])['Value'].mean().sort_values(ascending=False)
print("Highest Region Coverage in 2008: ", region_2008_avg.index[0],"with a coverage of",region_2008_avg.head(1).values[0])
print("Highest Region Coverage in 2014: ", region_2014_avg.index[0],"with a coverage of", region_2014_avg.head(1).values[0])

#Lowest in each year
lowest_2008_avg = region_2008_avg.tail(1)
lowest_2014_avg = region_2014_avg.tail(1)

print("Lowest Region Coverage in 2008: ", lowest_2008_avg.index[0],"with a coverage of", lowest_2008_avg.values[0])
print("Lowest Region Coverage in 2014: ", lowest_2014_avg.index[0], "with a coverage of", lowest_2014_avg.values[0])
```

```python
#Compare if there's an increase/decrease in the highest and lowest region coverage per year
highest_2008_avg = region_2008_avg.head(1).values[0]
highest_2014_avg = region_2014_avg.head(1).values[0]
lowest_2008_avg = region_2008_avg.tail(1).values[0]
lowest_2014_avg = region_2014_avg.tail(1).values[0]

#2008 - 2014 lowest
if lowest_2014_avg > lowest_2008_avg :
  lowest_diff = lowest_2014_avg - lowest_2008_avg
  print("Lowest coverage region increased in 2014 by" , round(lowest_diff,2))
  perc_inc = (lowest_diff/lowest_2014_avg) * 100
  print("Percentage increase: ", round(perc_inc,2),"%")
else:
  lowest_diff = lowest_2008_avg - lowest_2014_avg
  print("Lowest coverage region decreased in 2014 by", round(lowest_diff,2))

  perc_dec = (lowest_diff/lowest_2008_avg) * 100
  print("Percentage decrease in lowest coverage: ", round(perc_dec,2),"%")

#2008 - 2014 highest
if highest_2014_avg > highest_2008_avg :
  highest_diff = highest_2014_avg - highest_2008_avg
  print("Highest coverage region increased in 2014 by" , round(highest_diff,2))

  perc_inc = (highest_diff/highest_2014_avg) * 100
  print("Percentage increase: ", round(perc_inc,2),"%")
else:
  highest_diff = highest_2008_avg - highest_2014_avg
  print("Lowest coverage decreased in 2014 by", round(highest_diff,2))

  perc_dec = (highest_diff/highest_2008_avg) * 100
  print("Percentage decrease in highest coverage: ", round(perc_dec,2),"%")

# Rift Valley increased by 40.22%
# North Eastern increased by 65.01%
```

##Indicator, Region by Insurance

```python
df['Indicator'].unique()
```

```python
df.groupby(['Indicator'])['Value'].mean().sort_values(ascending=False)
```

```python
indicator_interest=["No health insurance [Women]","No health insurance [Men]"]

#subset data to include the interested indicators
df_interest = df[df['Indicator'].isin(indicator_interest)]
df_interest_pivot = df_interest[df_interest['Indicator'].isin(indicator_interest)].pivot_table(
    values = 'Value',
    index =['SurveyYear'],
    columns = ['Indicator','Location'],
    aggfunc = 'sum'
).reset_index()
df_interest_pivot
```

```python
top_locations = []
# Loop through years
for year in df_interest['SurveyYear'].unique():
    for gender in indicator_interest:
        sub_df = df_interest[(df_interest['SurveyYear'] == year) & (df_interest['Indicator'] == gender)]
        max_row = sub_df.loc[sub_df['Value'].idxmax()]
        top_locations.append({
            'Year': year,
            'Gender': gender.split("[")[-1].replace("]", ""),
            'Location': max_row['Location'],
            'Value': max_row['Value']
        })

# Convert to DataFrame
top_locations_df = pd.DataFrame(top_locations)
display(top_locations_df)

#Increase in value for women with no insurance cover
# No change in value for men with no insurance cover
```

```python
indicators=["Privately purchased commercial insurance [Women]","Privately purchased commercial insurance [Men]","Social security health insurance [Women]","Social security health insurance [Men]","Other employer-base health insurance [Women]","Other employer-base health insurance [Men]","Mutual health organzation or community-base health insurance [Women]","Mutual health organzation or community-base health insurance [Men]"]
focus_df=df[df["Indicator"].isin(indicators)]
focus_df
```
